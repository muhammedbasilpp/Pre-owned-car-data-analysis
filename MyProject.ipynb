## Data Cleaning

- Remove text units and convert to numbers

> Changing datatypes

```python
df['engine'] = pd.to_numeric(
    df['engine']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None),
    errors='coerce'
)
df
```

```python
df['mileage'] = pd.to_numeric(
    df['mileage']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None)
)
df
```

```python
df['max_power'] = pd.to_numeric(
    df['max_power']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None)
)
df
```

## Filling missing values

1.Engine rows filling through median method

```python
df['engine']=df['engine'].fillna(df['engine'].median())
```

2.Mileage rows filling through median

```python
df['mileage']=df['mileage'].fillna(df['mileage'].median())
```

3. Max power rows filling through median method

```python
df['max_power']=df['max_power'].fillna(df['max_power'].median())
```

4.seats rows filling though making 5 

```python
df['seats']=df['seats'].fillna(5)
```
