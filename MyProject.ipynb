## Data Cleaning

- Remove text units and convert to numbers

> Changing datatypes

```python
df['engine'] = pd.to_numeric(
    df['engine']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None),
    errors='coerce'
)
df
```

```python
df['mileage'] = pd.to_numeric(
    df['mileage']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None)
)
df
```

```python
df['max_power'] = pd.to_numeric(
    df['max_power']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None)
)
df
```

- Categorical data statistics summary

```python
df.describe(include=object)
```

1. Fuel 
- There are only 4 unique fuel types, but "Diesel" is the most frequent (3,755 occurrences).
> Insight: More than 54% of the cars in this dataset run on Diesel. This suggests a market preference or a dataset heavily skewed toward larger or commercial-style passenger vehicles.

2. Seller type
- With only 3 unique types, "Individual" sellers make up the vast majority (6,218 out of 6,926).

> Insight: Approximately 90% of the listings are from private individuals rather than dealers. You are looking at a peer-to-peer heavy marketplace.

3. Transmission
- There are only 2 unique values (likely Manual and Automatic). "Manual" is the overwhelming leader with 6,342 entries.

> Insight: Over 91% of the cars have manual transmissions. This indicates the dataset likely comes from a market where manual cars are the standard or more affordable option.

4. owner
- There are 5 unique owner categories. "First Owner" is the most common (4,242).

> Insight: About 61% of the cars are being sold by their original owners. This usually implies the vehicles might be in better condition compared to those with multiple previous owners.

5. Name
- There are 2,058 unique car names, which is very high relative to the total count.

- The Maruti Swift Dzire VDI is the most frequent model, appearing 118 times.

> Insight: The market is highly fragmented with many different models, but the Swift Dzire's top spot suggests it is one of the most popular and frequently traded used cars in this region.
